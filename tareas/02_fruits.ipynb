```python
order_frecuency= {}

for fruit in fruits:
    order = fruit["order"]
    if order in order_frecuency:
        order_frecuency[order] += 1
    else:
        order_frecuency[order] = 1

max_frecuency = 0
max_order = ""
for order_item in order_frecuency.items():
# for (order, frecuency) in order_frecuency.items():
    (order, frecuency) = order_item
    # order = order_item[0]
    # frecuency = order_item[1]
    if frecuency > max_frecuency:
        max_frecuency = frecuency
        max_order = order

total_nutritions = {
    "calories": 0,
    "fat": 0,
    "sugar": 0,
    "carbohydrates": 0,
    "protein": 0,
}
for fruit in fruits:
    if fruit["order"] == max_order:
        nutritions = fruit["nutritions"]
        total_nutritions["calories"] += nutritions["calories"]
        total_nutritions["fat"] += nutritions["fat"]
        total_nutritions["sugar"] += nutritions["sugar"]
        total_nutritions["carbohydrates"] += nutritions["carbohydrates"]
        total_nutritions["protein"] += nutritions["protein"]


print(f"El orden con mas frutas es {max_order} con {max_frecuency} frutas.")
for nutritions, value in total_nutritions.items():
    print(f"El total de {nutritions} es: {value}")
```
